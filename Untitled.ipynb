```python
import pandas as pd
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
'lat':[25.206570, 21.951734,25.158249],
'lon':[121.690427,120.781248,121.768227],
'name':['Yehliu','Kenting','Heping']
})
```

```python
list(data.lon)
```

```python
plt.figure(figsize=(10,20))

m = Basemap(projection='cyl',llcrnrlat=21.5,urcrnrlat=25.5,\
            llcrnrlon=119.5,urcrnrlon=122.5,resolution='i',epsg=3825)
m.arcgisimage(service='World_Shaded_Relief', xpixels = 400)

m.drawcoastlines()
x, y = m(list(data.lon), list(data.lat))
m.scatter(x,y,s=100)
i = 0
for xx,yy in zip(x,y):
    if i == 2:
        plt.text(xx+7000,yy-10000,data.name[i],fontsize=20)
    else:
        plt.text(xx+7000,yy+50,data.name[i],fontsize=20)
    i += 1
plt.savefig("taiwan_clustering.png",bbox_inches="tight")
plt.show()
```
