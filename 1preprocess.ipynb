```python
from glob import glob
from IPython.display import Image as IImage, display
from exif import Image
from resizeimage import resizeimage
import PIL
```

```python
def dms_dd(coor):
    degrees, minutes, seconds = coor
    if degrees >= 0:
        decimal = degrees + minutes/60.0 + seconds/3600.0
    else:
        decimal = degrees - minutes/60.0 - seconds/3600.0
    return decimal
```

```python
!rm -rf flickr/yehliu/preprocess
!mkdir -p flickr/yehliu/preprocess
for i,file in enumerate(glob("flickr/yehliu/origin/*jpg")):
    print(file)
    with open(file, 'rb') as image_file:
        my_image = Image(image_file)
    image = PIL.Image.open(file)
    cover = resizeimage.resize_cover(image, [448, 448])
    try:
        print(my_image.datetime)
        exif = image.info['exif']
        cover.save(file.split("/")[0] + "/" + file.split("/")[1] + "/preprocess/" + file.split("/")[-1],exif=exif)
    except:
        pass
!ls flickr/yehliu/preprocess
```

```python
for file in glob("flickr/yehliu/preprocess/*jpg"):
    with open(file, 'rb') as image_file:
        my_image = Image(image_file)
        print(my_image.has_exif)
        if my_image.has_exif:
            try:
                print(dms_dd(my_image.gps_latitude), dms_dd(my_image.gps_longitude), my_image.datetime)
            except:
                print(my_image.datetime)
```
