```python
loc = "kenting"
keyword = 'kenting park ocean'

photos = flickr.walk(text=keyword,
                     tag_mode='all',
                     extras='url_c',
#                     tags='-people,-human',
                     per_page=100,           # may be you can try different numbers..
                     sort='relevance')

urls = []
for i, photo in enumerate(photos):
    url = photo.get('url_c')
    urls.append(url)    
    # get 50 urls
    if i > 300:
        break
```
